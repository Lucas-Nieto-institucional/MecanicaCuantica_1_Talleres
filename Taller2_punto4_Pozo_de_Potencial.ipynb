# 4) Pozo infinito de potencial

Una partícula se encuentra dentro de un pozo de potencial infinito de la forma:

$$V(x) = \begin{cases}
          0, & 0 \le x \le a \\
          \infty, & \text{de lo contrario.}\\
         \end{cases}$$
    
Con una función de onda inicial dada por:

$$\Psi(x,0) = Ax(a-x), \;\; (0 \le x \le a)$$

a) Encontrar los coeficientes $c_n$ de $\Psi(x,0)$ en términos de las soluciones del pozo de potencial infinito $\psi(x)$. Verificar que se cumple que: 

$$ \int_{-\infty}^{\infty} \| \Psi(x,0) \| ^{2} \,dx = \sum_{n=1}^{\infty} \| c_n \| ^{2} = 1$$

#### Solución:
Tenemos que en el interior del pozo, dado que el potencial es nulo, la parte espacial de la ecuación de Schrödinger toma la forma canónica de un oscilador armónico cuyas soluciones serán:

$$\psi(x) = Ae^{ikx} + Be^{-ikx}$$

Aplicando condiciones de continuidad $(\psi(0) = \psi(a) = 0)$ hallamos que $B=-A$ y $k=\frac{n\pi}{a}$, lo que nos permite escribir la solución del pozo de potencial infinito como:

$$\psi(x) = \sqrt{\frac{2}{a}}\sin(\frac{n\pi x}{a})$$

Por la relación de completés de las soluciones $\psi_n(x)$, cualquier función $f(x)$ puede ser expandida en términos de dichas soluciones, de forma que:

$$ f(x) = \sum_{n=1}^{\infty} c_n \psi_n(x) $$

Donde usando la propiedad de ortonormalidad de $\psi_n(x)$:

$$\int \psi_m^{*}(x) \psi_n(x) \,dx = \delta_{mn}$$

Los coeficientes $c_n$ vienen dados por:

$$ c_n = \int \psi_n^{*}(x) f(x) \,dx$$

Entonces, al considerar $f(x) = \Psi(x,0)$, se debe poder expresar esta función de onda inicial en términos de las funciones propias del pozo de potencial:

$$Ax(a−x) = \sum_{n=1}^{\infty} c_n \sqrt{\frac{2}{a}}\sin(\frac{n\pi x}{a})$$

Dado que estamos buscando aplicar las propiedades de ortonormalidad, integramos $\psi_m(x)$ (no nos preocupamos por conjugar $\psi(x)$, pues es una función real) a ambos lados de la igualdad:

$$\sqrt{\frac{2}{a}}\int_{0}^{a} Ax(a−x)\sin(\frac{n\pi x}{a})\,dx = \int_{0}^{a} \sqrt{\frac{2}{a}}\sin(\frac{m\pi x}{a})\sum_{n=1}^{\infty} c_n \sqrt{\frac{2}{a}}\sin(\frac{n\pi x}{a})\,dx$$

$$\sqrt{\frac{2}{a}}\int_{0}^{a} Ax(a−x)\sin(\frac{m\pi x}{a})\,dx = \frac{2}{a}\sum_{n=1}^{\infty} c_n\int_{0}^{a} \sin(\frac{m\pi x}{a}) \sin(\frac{n\pi x}{a})\,dx$$

$$\sqrt{\frac{2}{a}}\int_{0}^{a} Ax(a−x)\sin(\frac{m\pi x}{a})\,dx = \frac{2}{a}\sum_{n=1}^{\infty} c_n \delta_{mn}$$

$$\sqrt{\frac{2}{a}}\int_{0}^{a} Ax(a−x)\sin(\frac{m\pi x}{a})\,dx = \frac{2}{a}c_m$$

$$c_m =  \frac{Aa}{2}\sqrt{\frac{2}{a}}\int_{0}^{a} x(a−x)\sin(\frac{m\pi x}{a})\,dx$$

Para hallar $c_n$ debemos resolvemos la integral:

$$c_n =  \frac{Aa}{2}\sqrt{\frac{2}{a}}\int_{0}^{a} x(a−x)\sin(\frac{n\pi x}{a})\,dx$$

$$c_n =  \frac{Aa}{2}\sqrt{\frac{2}{a}} \left( -\frac{a^{3}\left( n\pi\sin(n\pi) + 2\cos(n\pi)-2\right)}{n^3 \pi^3} \right) $$

$$c_n =  \frac{Aa}{2}\sqrt{\frac{2}{a}} \left( -\frac{a^{3}\left(2\cos(n\pi)-2\right)}{n^3 \pi^3} \right) $$

$$c_n =  \frac{Aa}{2}\sqrt{\frac{2}{a}} \left( -\frac{2a^{3}\left(2\cos(n\pi)-2\right)}{n^3 \pi^3} \right) $$

$$c_n =  \sqrt{\frac{2}{a}} \left( -\frac{A\;a^{4}\left(2\cos(n\pi)-2\right)}{n^3 \pi^3} \right) $$

$$c_n =  \frac{-A\sqrt{2}\;a^{\frac{5}{2}}\left(2\cos(n\pi)-2\right)}{n^3 \pi^3} $$

Únicamente resta hallar el valor de $A$, para ello sabemos que la función de onda $\Psi(x,0)$ debe estar normalizada:

$$ \int_{-\infty}^{\infty} \| \Psi(x,0) \| ^{2} \,dx = 1 $$

Es decir:

$$ \int_{-\infty}^{\infty} \Psi(x,0)\Psi^{*}(x,0) \,dx = 1 $$

Pero dado que $\Psi(x,0)$ es real, se cumple que $\Psi(x,0)=\Psi^{*}(x,0)$, entonces:

$$ \int_{0}^{a} Ax(a-x)\;Ax(a-x) \,dx = 1 $$

$$ A^2\int_{0}^{a} x^2(a-x)^2 \,dx = 1 $$

$$ A^2\int_{0}^{a} x^2(a^2 - 2ax +x^2) \,dx = 1 $$

$$ A^2\int_{0}^{a} (a^2x^2 - 2ax^3 +x^4) \,dx = 1 $$

$$ A^2\left(\int_{0}^{a}a^2x^2\,dx - \int_{0}^{a}2ax^3\,dx +\int_{0}^{a}x^4\,dx\right)  = 1 $$

$$ A^2\left(a^2\int_{0}^{a}x^2\,dx - 2a\int_{0}^{a}x^3\,dx +\int_{0}^{a}x^4\,dx\right)  = 1 $$

$$ A^2\left(a^2\frac{a^3}{3}- 2a\frac{a^4}{4} +\frac{a^5}{5}\right)  = 1 $$

$$ A^2\left(\frac{a^5}{3}- \frac{a^5}{2} +\frac{a^5}{5}\right)  = 1 $$

$$ a^5A^2\left(\frac{1}{3}- \frac{1}{2} +\frac{1}{5}\right)  = 1 $$

$$ \frac{a^5A^2}{30}  = 1 $$

$$ A = \sqrt{\frac{30}{a^5}} $$

Finalmente, $c_n$ será: 

$$c_n = \frac{-\sqrt{\frac{30}{a^5}}\sqrt{2}\;a^{\frac{5}{2}}\left(2\cos(n\pi)-2\right)}{n^3 \pi^3} $$

$$c_n = \frac{-\sqrt{30}\sqrt{2}\;a^{\frac{5}{2}}\left(2\cos(n\pi)-2\right)}{\sqrt{a^5}n^3 \pi^3} $$

$$c_n = \frac{-\sqrt{30}\sqrt{2}\;a^{\frac{5}{2}}\left(2\cos(n\pi)-2\right)}{a^{\frac{5}{2}}n^3 \pi^3} $$

$$c_n = \frac{-\sqrt{60}\;\left(2\cos(n\pi)-2\right)}{n^3 \pi^3} $$

$$c_n = \frac{-2\sqrt{15}\left(2\cos(n\pi)-2\right)}{n^3 \pi^3} $$

$$c_n = \frac{-4\sqrt{15}\left(\cos(n\pi)-1\right)}{n^3 \pi^3} $$

$$c_n = \frac{-4\sqrt{15}\left((-1)^n-1\right)}{n^3 \pi^3} $$

$\;$

```python
# Demostremos que La suma del módulo cuadrado de c_n también converge a 1:

from scipy.constants import pi
import numpy as np
n_terms = 10
convergence = 0

def c_n(n):
    term = ((-4*np.sqrt(15))*(((-1)**n)-1))/((n*pi)**3)
    return term

for i in range(1,n_terms):
    convergence += (c_n(i))**2
    print(convergence)
    
print(f'Sólo con los primeros {n_terms} términos, la suma converge a {round(convergence)}')
```

b) Escriba un programa y haga gráficas de las funciones de onda, incluyendo la parte
temporal, para diversos valores de n y para las distribuciones de probabilidad asociadas. Puede
usar valores $a$ y $t$ que considera apropiados.

#### Solución:

```python
from scipy.constants import hbar
from scipy.constants import m_e # Masa del electrón, para simular un electrón en un pozo infinito de potencial jeje

def sol_temporal(n,a,t,m = m_e):
    # Solución temporal (real) a la ecuación de Schrödinger demostrada en clase, pero dado que el potencial es independiente
    # Del tiempo, la densidad de probabilidad asociada también es independiente del tiempo.
    # Cuantización de energía:
    E = ((n*pi*hbar)**2)/(2*m*(a**2)) # Ecuación 2.30 de Introduction to quantum mechanics 3rd ed (Griffiths., Schroeter)
    factor = -E/hbar
    return np.cos(factor*t) - np.sin(factor*t)

def sol_espacial(n,a,x):
    return np.sqrt(2/a)*np.sin((n*pi*x)/(a))

def psi(n,a,x,t):
    # Dado que para el potencial constante se puede asumir separación de variables:
    return sol_espacial(n,a,x)*sol_temporal(n,a,t)

def Prob_density(psi):
    return psi*np.conjugate(psi)
```

```python
import matplotlib.pyplot as plt
a = 2*pi
x_arr = np.linspace(0,a,100)

fig, axs = plt.subplots(1,3, figsize=(10, 3))
fig.suptitle("Función de onda, densidad de probabilidad y evolución temporal en pozo infinito 1D en el primer nivel de energía")
fig.tight_layout(rect=[0, 0, 1, 1])

for ax in axs:
        ax.set_xlim([-1/2, 2.1*pi])
        ax.set_ylim([-.75, .75])
        ax.grid()
psi_0 = psi(1,a,x_arr,0) # t = 0
psi_1 = psi(1,a,x_arr,1) # t = 1
p = Prob_density(psi_0)
axs[0].set_title('\n'+r'$\psi_1(x,0)$')
axs[0].plot(x_arr,psi_0,c='r')
axs[1].set_title('\n'+r'$\rho_1(x,0)$')
axs[1].plot(x_arr,p,c='r')
axs[2].set_title('\n'+r'$\psi_1(x,1)$')
axs[2].plot(x_arr,psi_1,c='r')

plt.show()

fig, axs = plt.subplots(1,3, figsize=(10, 3))
fig.suptitle("Función de onda, densidad de probabilidad y evolución temporal en pozo infinito 1D en el segundo nivel de energía")
fig.tight_layout(rect=[0, 0, 1, 1])

for ax in axs:
        ax.set_xlim([-1/2, 2.1*pi])
        ax.set_ylim([-.75, .75])
        ax.grid()
psi_0 = psi(2,a,x_arr,0) # t = 0
psi_1 = psi(2,a,x_arr,1) # t = 1
p = Prob_density(psi_0)
axs[0].set_title('\n'+r'$\psi_2(x,0)$')
axs[0].plot(x_arr,psi_0,c='g')
axs[1].set_title('\n'+r'$\rho_2(x,0)$')
axs[1].plot(x_arr,p,c='g')
axs[2].set_title('\n'+r'$\psi_2(x,1)$')
axs[2].plot(x_arr,psi_1,c='g')

plt.show()

fig, axs = plt.subplots(1,3, figsize=(10, 3))
fig.suptitle("Función de onda, densidad de probabilidad y evolución temporal en pozo infinito 1D en el tercer nivel de energía")
fig.tight_layout(rect=[0, 0, 1, 1])

for ax in axs:
        ax.set_xlim([-1/2, 2.1*pi])
        ax.set_ylim([-.75, .75])
        ax.grid()
psi_0 = psi(3,a,x_arr,0) # t = 0
psi_1 = psi(3,a,x_arr,1) # t = 1
p = Prob_density(psi_0)
axs[0].set_title('\n'+r'$\psi_3(x,0)$')
axs[0].plot(x_arr,psi_0,c='b')
axs[1].set_title('\n'+r'$\rho_3(x,0)$')
axs[1].plot(x_arr,p,c='b')
axs[2].set_title('\n'+r'$\psi_3(x,1)$')
axs[2].plot(x_arr,psi_1,c='b')

plt.show()
```
